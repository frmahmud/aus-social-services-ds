## 1. Background & Motivation

The Australian Department of Social Services (DSS) delivers welfare payments and 
support services to millions of Australians. Demand for these services is uneven 
across regions and shifts over time due to economic conditions, population changes, 
and policy decisions.

Currently, demand forecasting is largely reactive. This project asks:

> **Can we use publicly available data to proactively identify which communities 
> face the highest risk of increased social services demand?**

This has direct policy value — early identification of high-risk regions allows 
government to allocate resources, staff services, and design interventions ahead 
of demand spikes rather than after them.

## 2. Research Questions

This project answers three questions, each corresponding to a distinct DS technique:

| # | Question | Technique |
|---|----------|-----------|
| 1 | Which SA2 regions are at high / medium / low risk of welfare demand growth? | Classification (ML) |
| 2 | How will social service demand trend over the next 6–12 months? | Time-series forecasting |
| 3 | What themes dominate recent Australian social policy documents? | NLP / Topic modelling |

````python
# This cell documents our research questions as Python variables
# so they can be referenced consistently throughout the project

RQ1 = "Classify SA2 regions as high/medium/low risk for welfare demand growth"
RQ2 = "Forecast quarterly social service demand 6-12 months ahead"
RQ3 = "Extract dominant topics from Australian social policy documents (NLP)"

print("Research Questions Defined:")
print(f"  RQ1: {RQ1}")
print(f"  RQ2: {RQ2}")
print(f"  RQ3: {RQ3}")
````

## 4. Success Metrics

Before building any model, we define what "good" looks like. 
This prevents moving the goalposts after seeing results.

````python
# Define success metrics for each research question

success_metrics = {
    "RQ1 — Classification": {
        "primary_metric"  : "F1 Score (macro average)",
        "target"          : ">= 0.70",
        "why"             : "Macro F1 treats each class equally — important because high-risk regions are rare (imbalanced classes)",
        "secondary_metric": "ROC-AUC, Precision-Recall curve"
    },
    "RQ2 — Forecasting": {
        "primary_metric"  : "MAPE (Mean Absolute Percentage Error)",
        "target"          : "<= 10%",
        "why"             : "Percentage error is interpretable for policy stakeholders — '8% error on forecast' is meaningful",
        "secondary_metric": "MAE, RMSE, visual forecast vs actuals chart"
    },
    "RQ3 — NLP": {
        "primary_metric"  : "Topic coherence score (C_v)",
        "target"          : ">= 0.55",
        "why"             : "Coherence measures whether discovered topics make semantic sense — higher = more interpretable topics",
        "secondary_metric": "Manual review of top 10 words per topic, perplexity score"
    }
}

for rq, metrics in success_metrics.items():
    print(f"{'='*55}")
    print(f"  {rq}")
    print(f"{'='*55}")
    print(f"  Primary metric : {metrics['primary_metric']}")
    print(f"  Target         : {metrics['target']}")
    print(f"  Why            : {metrics['why']}")
    print(f"  Secondary      : {metrics['secondary_metric']}")
    print()
````

## 5. Ethics & Privacy Checklist

Working in the Australian Public Sector requires explicit consideration of 
ethical obligations before collecting or analysing any data. This checklist 
is aligned with:

- **Privacy Act 1988** (Cth) and the Australian Privacy Principles (APPs)
- **APS AI Ethics Principles** (DISER, 2024)
- **DSS Data Governance Framework**

````python
# Ethics and privacy checklist
# Every item should be TRUE before proceeding to data collection

ethics_checklist = [
    {
        "item"  : "Only publicly available, open-licensed data is used",
        "status": True,
        "notes" : "All sources are ABS, DSS, and AIHW open data — no internal DSS systems accessed"
    },
    {
        "item"  : "No individual-level personal information is collected",
        "status": True,
        "notes" : "All data is aggregated at SA2 region level — no names, TFNs, or CRNs"
    },
    {
        "item"  : "Data does not identify specific vulnerable individuals",
        "status": True,
        "notes" : "SA2 regions contain 3,000–25,000 people — aggregation prevents re-identification"
    },
    {
        "item"  : "Model outputs will not be used for automated decision-making about individuals",
        "status": True,
        "notes" : "Project is for portfolio / research purposes only — not a production decision system"
    },
    {
        "item"  : "Indigenous community data handled with additional care",
        "status": True,
        "notes" : "SEIFA and DSS data includes remote/Indigenous SA2s — results will be interpreted with cultural context, not used to stigmatise communities"
    },
    {
        "item"  : "Data will not be re-published or redistributed",
        "status": True,
        "notes" : "Raw data excluded from GitHub via .gitignore — only analysis outputs shared"
    },
    {
        "item"  : "Model limitations and potential biases will be documented",
        "status": True,
        "notes"  : "A MODEL_CARD.md will be completed in Phase 7 before GitHub publication"
    },
]

all_clear = all(item["status"] for item in ethics_checklist)

print("ETHICS & PRIVACY CHECKLIST")
print("="*55)
for item in ethics_checklist:
    tick = "✓" if item["status"] else "✗"
    print(f"\n  [{tick}] {item['item']}")
    print(f"       {item['notes']}")

print("\n" + "="*55)
if all_clear:
    print("  ✓ All checks passed — cleared to proceed to Phase 2")
else:
    print("  ✗ One or more checks failed — review before proceeding")
````
