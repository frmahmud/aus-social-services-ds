```python
## 2. Download Helper Function

Rather than downloading files manually, we write a reusable function that:
- Checks if the file already exists before downloading (avoids re-downloading)
- Shows file size after download
- Handles network errors gracefully

This is a standard pattern in professional data pipelines.
```

```python
def download_file(url, destination, description=""):
    """
    Download a file from a URL to a local destination path.
    Skips download if the file already exists.
    
    Parameters
    ----------
    url         : str  — the download URL
    destination : str  — full local file path to save to
    description : str  — human-readable label for progress output
    
    Returns
    -------
    bool — True if file is available (downloaded or already existed)
    """
    if os.path.exists(destination):
        size_mb = os.path.getsize(destination) / (1024 * 1024)
        print(f"  [✓] Already exists — {description} ({size_mb:.1f} MB)")
        return True

    print(f"  [↓] Downloading — {description} ...")
    
    try:
        # stream=True downloads in chunks — important for large files
        # so we don't load the entire file into RAM before saving
        response = requests.get(url, stream=True, timeout=60)
        response.raise_for_status()   # raises an error if HTTP 4xx/5xx
        
        with open(destination, "wb") as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        
        size_mb = os.path.getsize(destination) / (1024 * 1024)
        print(f"  [✓] Saved to {destination} ({size_mb:.1f} MB)")
        return True

    except requests.exceptions.RequestException as e:
        print(f"  [✗] Download failed: {e}")
        print(f"       Please download manually from: {url}")
        return False


print("download_file() function defined and ready")
```

## 3. Dataset 1 — ABS SEIFA 2021

**What it is:** The Socio-Economic Indexes for Areas (SEIFA) ranks every SA2 
region in Australia by four indexes:

- **IRSD** — Index of Relative Socio-economic Disadvantage (most commonly used)
- **IRSAD** — Index of Relative Socio-economic Advantage and Disadvantage  
- **IER** — Index of Economic Resources  
- **IEO** — Index of Education and Occupation

A low IRSD score = high disadvantage. We will use IRSD as a key feature 
in our classification model.

**Source:** ABS — abs.gov.au

```python
# ABS SEIFA 2021 — SA2 level
# Direct download link from ABS website

SEIFA_URL  = "https://www.abs.gov.au/statistics/people/people-and-communities/socio-economic-indexes-areas-seifa-australia/2021/Statistical%20Area%20Level%202%2C%20Indexes%2C%20SEIFA%202021.xlsx"
SEIFA_FILE = os.path.join(RAW_DIR, "seifa_2021.xlsx")

print("Dataset 1 — ABS SEIFA 2021")
print("-" * 45)
success = download_file(SEIFA_URL, SEIFA_FILE, "ABS SEIFA 2021 SA2")

if not success:
    print()
    print("  Manual download steps:")
    print("  1. Go to: https://www.abs.gov.au/statistics/people/people-and-communities/socio-economic-indexes-areas-seifa-australia/latest-release")
    print("  2. Download: 'Statistical Area Level 2, Indexes, SEIFA 2021.xlsx'")
    print(f"  3. Save to: {SEIFA_FILE}")
```

```python
## 4. Load & Inspect SEIFA Data

Once downloaded, we load the Excel file, inspect its structure, 
and extract the sheet we need. ABS Excel files typically have 
multiple sheets and metadata rows at the top — we need to handle this.
```

```python
# First, inspect what sheets are available in the Excel file

if os.path.exists(SEIFA_FILE):
    xl = pd.ExcelFile(SEIFA_FILE)
    print("Sheets available in SEIFA file:")
    for i, sheet in enumerate(xl.sheet_names):
        print(f"  {i}: {sheet}")
else:
    print("File not yet downloaded — run Cell 8 first")
```

```python
# Load the Table 2 sheet which contains SA2-level IRSD scores
# skiprows=5 skips the ABS metadata header rows at the top of the file
# This number may vary — if you get an error, try skiprows=4 or skiprows=6

if os.path.exists(SEIFA_FILE):
    df_seifa = pd.read_excel(
        SEIFA_FILE,
        sheet_name="Table 2",   # SA2 level data
        skiprows=5,             # Skip ABS metadata header rows
        engine="openpyxl"
    )

    # Drop completely empty rows and columns (common in ABS Excel files)
    df_seifa = df_seifa.dropna(how="all").dropna(axis=1, how="all")

    print(f"SEIFA loaded:  {df_seifa.shape[0]:,} rows  x  {df_seifa.shape[1]} columns")
    print()
    print("First 5 rows:")
    df_seifa.head()
else:
    print("File not found — please complete download step first")
```

```python
# Rename columns to clean, consistent names
# Column names from ABS files are often long and inconsistent

if os.path.exists(SEIFA_FILE):
    
    # Print current column names so we know what we're working with
    print("Current column names:")
    for i, col in enumerate(df_seifa.columns):
        print(f"  {i}: {col}")
```

```python
# Standardise column names based on the ABS SEIFA 2021 structure
# Adjust these if your column names differ slightly

if os.path.exists(SEIFA_FILE):
    df_seifa.columns = df_seifa.columns.str.strip()

    rename_map = {}
    for col in df_seifa.columns:
        col_lower = str(col).lower().strip()
        if "sa2" in col_lower and "code" in col_lower:
            rename_map[col] = "sa2_code"
        elif "sa2" in col_lower and "name" in col_lower:
            rename_map[col] = "sa2_name"
        elif "irsd" in col_lower and "score" in col_lower:
            rename_map[col] = "irsd_score"
        elif "irsd" in col_lower and "decile" in col_lower:
            rename_map[col] = "irsd_decile"
        elif "irsad" in col_lower and "score" in col_lower:
            rename_map[col] = "irsad_score"
        elif "irsad" in col_lower and "decile" in col_lower:
            rename_map[col] = "irsad_decile"
        elif "usual" in col_lower or "population" in col_lower:
            rename_map[col] = "usual_resident_population"
        elif "state" in col_lower:
            rename_map[col] = "state_name"

    df_seifa = df_seifa.rename(columns=rename_map)

    print("Renamed columns:")
    print(df_seifa.columns.tolist())
    print()
    df_seifa.head(3)
```

```python
# Data quality check on SEIFA

if os.path.exists(SEIFA_FILE):
    print("SEIFA — Data Quality Report")
    print("=" * 40)
    print(f"  Rows              : {len(df_seifa):,}")
    print(f"  Columns           : {df_seifa.shape[1]}")
    print()
    print("  Missing values per column:")
    missing = df_seifa.isnull().sum()
    for col, count in missing.items():
        pct = count / len(df_seifa) * 100
        flag = "  ← check this" if pct > 5 else ""
        print(f"    {col:<35} {count:>5}  ({pct:.1f}%){flag}")
    print()
    print("  Data types:")
    print(df_seifa.dtypes)
```

```python
# Save a clean version of SEIFA to processed folder

if os.path.exists(SEIFA_FILE):
    seifa_out = os.path.join(PROCESSED_DIR, "seifa_clean.csv")
    df_seifa.to_csv(seifa_out, index=False)
    print(f"[✓] SEIFA clean data saved → {seifa_out}")
    print(f"    Shape: {df_seifa.shape}")
```

## 5. Dataset 2 — DSS Payment Demographic Data

**What it is:** Quarterly counts of DSS payment recipients broken down by:
- Payment type (JobSeeker, Age Pension, Disability Support, etc.)
- SA2 region
- Age group and gender

This is the primary **target variable source** for our classification 
and forecasting models. The quarterly change in recipient counts per SA2 
is what we are trying to predict.

**Source:** data.gov.au — published by DSS

```python
# DSS Payment Demographic Data
# Available as quarterly CSV files on data.gov.au
# We will download the most recent available quarter

DSS_URL  = "https://data.gov.au/data/dataset/cff2ae8a-55e4-47db-a66d-e177fe0ac6a0/resource/857e9fc6-c539-4aa5-b4e8-cfc2b18cc3b5/download/dss-demographics-jun-2023.csv"
DSS_FILE = os.path.join(RAW_DIR, "dss_payments_jun2023.csv")

print("Dataset 2 — DSS Payment Demographic Data")
print("-" * 45)
success = download_file(DSS_URL, DSS_FILE, "DSS Payments Jun 2023")

if not success:
    print()
    print("  Manual download steps:")
    print("  1. Go to: https://data.gov.au/dataset/ds-dga-cff2ae8a-55e4-47db-a66d-e177fe0ac6a0")
    print("  2. Download the most recent quarterly CSV file")
    print(f"  3. Save to: {RAW_DIR}")
    print("  4. Update DSS_FILE path in this cell to match your filename")
```

```python
# Load and inspect DSS payment data

if os.path.exists(DSS_FILE):
    df_dss = pd.read_csv(DSS_FILE, encoding="utf-8-sig")
    # utf-8-sig handles the BOM character that Microsoft tools add to CSVs
    
    print(f"DSS data loaded: {df_dss.shape[0]:,} rows x {df_dss.shape[1]} columns")
    print()
    print("Column names:")
    for col in df_dss.columns:
        print(f"  {col}")
    print()
    df_dss.head()
else:
    print("File not found — please complete download step first")
```

```python
# Data quality check on DSS payments

if os.path.exists(DSS_FILE):
    print("DSS Payments — Data Quality Report")
    print("=" * 40)
    print(f"  Rows    : {len(df_dss):,}")
    print(f"  Columns : {df_dss.shape[1]}")
    print()

    print("  Unique values in key columns:")
    for col in df_dss.columns[:8]:
        n = df_dss[col].nunique()
        print(f"    {col:<40} {n:>6} unique values")

    print()
    print("  Missing values:")
    missing = df_dss.isnull().sum()
    missing = missing[missing > 0]
    if len(missing) == 0:
        print("    No missing values found")
    else:
        print(missing)
```

```python
# Standardise DSS column names to snake_case
# and save clean version

if os.path.exists(DSS_FILE):
    # Convert all column names to lowercase with underscores
    df_dss.columns = (
        df_dss.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("/", "_")
        .str.replace("-", "_")
        .str.replace("(", "")
        .str.replace(")", "")
    )

    print("Standardised column names:")
    for col in df_dss.columns:
        print(f"  {col}")

    dss_out = os.path.join(PROCESSED_DIR, "dss_payments_clean.csv")
    df_dss.to_csv(dss_out, index=False)
    print(f"\n[✓] DSS clean data saved → {dss_out}")
```

## 6. Dataset 3 — ABS Regional Population 2022

**What it is:** Estimated Resident Population (ERP) by SA2, 
broken down by age group and sex. 

We use this to:
- Normalise payment counts to **recipients per 1,000 population** 
  (removes the effect of region size)
- Create age-structure features (% population aged 65+, % aged 15–24)
  which are strong predictors of welfare demand type

**Source:** ABS — abs.gov.au

```python
# ABS Regional Population by SA2

POP_URL  = "https://www.abs.gov.au/statistics/people/population/regional-population/2021-22/32180DS0003_2001-22.xlsx"
POP_FILE = os.path.join(RAW_DIR, "abs_population_2022.xlsx")

print("Dataset 3 — ABS Regional Population 2022")
print("-" * 45)
success = download_file(POP_URL, POP_FILE, "ABS Regional Population 2022")

if not success:
    print()
    print("  Manual download steps:")
    print("  1. Go to: https://www.abs.gov.au/statistics/people/population/regional-population/latest-release")
    print("  2. Download: Table 3 — ERP by SA2")
    print(f"  3. Save to: {POP_FILE}")
```

```python
# Load and inspect population data

if os.path.exists(POP_FILE):
    xl_pop = pd.ExcelFile(POP_FILE)
    print("Sheets in population file:")
    for sheet in xl_pop.sheet_names:
        print(f"  {sheet}")
```

```python
# Load population data — adjust sheet name and skiprows if needed

if os.path.exists(POP_FILE):
    df_pop = pd.read_excel(
        POP_FILE,
        sheet_name=xl_pop.sheet_names[3],  # Usually Table 3
        skiprows=6,
        engine="openpyxl"
    )

    df_pop = df_pop.dropna(how="all").dropna(axis=1, how="all")

    print(f"Population data loaded: {df_pop.shape[0]:,} rows x {df_pop.shape[1]} columns")
    print()
    df_pop.head(3)
else:
    print("File not found — complete download step first")
```

```python
# Save population data

if os.path.exists(POP_FILE):
    pop_out = os.path.join(PROCESSED_DIR, "population_clean.csv")
    df_pop.to_csv(pop_out, index=False)
    print(f"[✓] Population data saved → {pop_out}")
    print(f"    Shape: {df_pop.shape}")
```

## 7. Merge Datasets on SA2 Code

Now we join SEIFA and DSS payment data on the SA2 code — 
the common geographic key across all ABS/DSS datasets.

This produces our **master analytical dataset** that later phases 
will use for EDA, feature engineering, and modelling.

```python
# Check both clean files exist before attempting merge

seifa_path = os.path.join(PROCESSED_DIR, "seifa_clean.csv")
dss_path   = os.path.join(PROCESSED_DIR, "dss_payments_clean.csv")

files_ready = True

for label, path in [("SEIFA clean", seifa_path), ("DSS clean", dss_path)]:
    if os.path.exists(path):
        print(f"  [✓] {label} found — {path}")
    else:
        print(f"  [✗] {label} NOT found — complete earlier steps first")
        files_ready = False

print()
if files_ready:
    print("  Both files present — ready to merge")
else:
    print("  One or more files missing — scroll up and complete download steps")
```

```python
# Merge SEIFA + DSS on SA2 code
# We use a left join on SEIFA so every SA2 region is kept,
# even if it has no DSS payment records (we will investigate those)

if files_ready:
    df_seifa_m = pd.read_csv(seifa_path)
    df_dss_m   = pd.read_csv(dss_path)

    # Standardise the SA2 code column to string for safe joining
    # SA2 codes are 9-digit integers but pandas sometimes reads as float
    if "sa2_code" in df_seifa_m.columns:
        df_seifa_m["sa2_code"] = df_seifa_m["sa2_code"].astype(str).str.strip().str.zfill(9)

    # Find the SA2 code column in DSS data (name varies by file version)
    dss_sa2_col = None
    for col in df_dss_m.columns:
        if "sa2" in col.lower() and ("code" in col.lower() or "2016" in col.lower() or "2021" in col.lower()):
            dss_sa2_col = col
            break

    if dss_sa2_col:
        df_dss_m[dss_sa2_col] = df_dss_m[dss_sa2_col].astype(str).str.strip().str.zfill(9)
        df_dss_m = df_dss_m.rename(columns={dss_sa2_col: "sa2_code"})

        df_master = pd.merge(
            df_seifa_m,
            df_dss_m,
            on="sa2_code",
            how="left"
        )

        print(f"Merge complete!")
        print(f"  SEIFA rows     : {len(df_seifa_m):,}")
        print(f"  DSS rows       : {len(df_dss_m):,}")
        print(f"  Master rows    : {len(df_master):,}")
        print(f"  Master columns : {df_master.shape[1]}")
        print()
        df_master.head(3)
    else:
        print("  [✗] Could not find SA2 code column in DSS data")
        print("  Available columns:", df_dss_m.columns.tolist())
```
